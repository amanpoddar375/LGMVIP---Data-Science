### Importing the libraries

```python
import matplotlib.pyplot as plt
import cv2
```

###  Reading the original image

```python
image=cv2.imread("priyankamohan.jpg")
plt.imshow(image)
plt.show()
```

### Converting the BGR image to RGB

```python
rgb_img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img)
plt.axis(False)
plt.show()
```

### Converting the RGB image to Gray Scale

```python
gray_img=cv2.cvtColor(rgb_img,cv2.COLOR_BGR2GRAY)
plt.imshow(gray_img)
```

### Inverting the gray scale image

```python
inverted_image= cv2.bitwise_not(gray_img)
plt.imshow(inverted_image)
```

### Blurring the image using Gaussian Function

```python
blur_img=cv2.GaussianBlur(inverted_image, (21, 21),0)
plt.imshow(blur_img)
```

### Mixing the Gray-Scale Image with Inverted Blurry Image

```python
invblur_img = cv2.bitwise_not(blur_img)
plt.imshow(invblur_img)
```

### Performing bit-wise division
The sketch can be generated by bit-wise division the grayscale image and the inverted-blurred image.

```python
sketch_img=cv2.divide(gray_img,invblur_img, scale=256.0)
```

### Saving the sketch image

```python
 cv2.imwrite('sketch.png', sketch_img)
```

### Displaying the original image and sketch side by side

```python
plt.figure(figsize=(14,8))
plt.subplot(1,2,1)
plt.title('Original image', size=18)
plt.imshow(rgb_img)
plt.axis('off')

plt.subplot(1,2,2)
plt.title('Sketch', size=18)
rgb_sketch=cv2.cvtColor(sketch_img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_sketch)
plt.axis('off')
plt.show()
```
